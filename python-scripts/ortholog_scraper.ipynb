```python
import re
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
url = "https://www.genome.jp/dbget-bin/get_linkdb?-t+genes+gn:T00015"
page = urlopen(url)
html = page.read().decode("utf-8")
soup = BeautifulSoup(html, "html.parser")
```

```python
x = soup.find_all("a")
x = x[1:len(x)]
```

```python
for item in x:
    ortholog = item.string
    print(ortholog[4:len(ortholog)])
```

```python
x = soup.find_all("pre")
```

```python
y = x[0]
```

```python
lst = str(y).split('\n')
lst = lst[2:len(lst)]
```

```python
count = 0
orthologs = []
for item in lst:
    start_idx = item.find(">mtu:")
    end_idx = item.find("</a")
    gene = item[start_idx+5:end_idx]
    match_result = re.search("K\d+",item)
    if match_result != None:
        KO_identifier = match_result.group()
        orthologs.append((gene,KO_identifier))
```
